## 9. Filter recipes by ingredients

We filter recipes based on relevant ingredients to keep only useful and realistic recipes.
The goal is to keep and use as much europan recipe more like Italian recepies. 

```python
italian_keywords = [
    "pasta", "tomato", "olive oil", "garlic",
    "basil", "parmesan", "mozzarella",
    "cheese", "chicken", "beef", "onion"
]

def has_relevant_ingredient(ingredients):
    ingredients = [i.lower() for i in ingredients]
    return any(
        keyword in ingredient
        for ingredient in ingredients
        for keyword in italian_keywords
    )
df = df[df["ingredients"].apply(has_relevant_ingredient)]
```

## Experimental filtering (European-style recipes)

We apply a stricter filtering rule based on ingredient similarity to European cuisine.

A recipe is kept only if most of its ingredients match a predefined keyword list.

```python
import pandas as pd
import ast
import json
from src.cookmate.utils.config import RAW_DATA_FILE, PROCESSED_DATA_DIR

chunk_size = 5000

clean_data = []
recipe_counter = 0

# expanded keyword list (european-style)
european_keywords = [
    "pasta", "spaghetti", "penne",
    "rice", "risotto",
    "potato", "potatoes",
    "tomato", "tomato sauce",
    "olive oil", "butter",
    "garlic", "onion",
    "carrot", "celery",
    "basil", "parsley",
    "cheese", "parmesan", "mozzarella", "ricotta",
    "milk", "cream",
    "egg", "eggs",
    "chicken", "beef", "pork",
    "sausage", "ham",
    "flour", "bread"
]

def safe_eval(x):
    try:
        return ast.literal_eval(x)
    except:
        return []

def european_score(ingredients):
    ingredients = [i.lower() for i in ingredients]
    
    total = len(ingredients)
    if total == 0:
        return 0
    
    matches = sum(
        any(keyword in ingredient for keyword in european_keywords)
        for ingredient in ingredients
    )
    
    return matches / total  # percentage


for chunk in pd.read_csv(RAW_DATA_FILE, chunksize=chunk_size):
    
    # select columns
    chunk = chunk[["title", "directions", "NER", "genre"]]
    
    # rename
    chunk = chunk.rename(columns={
        "directions": "instructions",
        "NER": "ingredients"
    })
    
    # filter genre
    chunk = chunk[~chunk["genre"].isin(["drinks", "fastfood", "fusion"])]
    
    # convert
    chunk["ingredients"] = chunk["ingredients"].apply(safe_eval)
    chunk["instructions"] = chunk["instructions"].apply(safe_eval)
    
    # apply european filter (STRICT)
    chunk = chunk[chunk["ingredients"].apply(lambda x: european_score(x) >= 0.9)]
    
    # build dataset
    for _, row in chunk.iterrows():
        recipe = {
            "id": f"recipe_{recipe_counter}",
            "title": row["title"],
            "ingredients": row["ingredients"],
            "instructions": row["instructions"]
        }
        clean_data.append(recipe)
        recipe_counter += 1
        
        if len(clean_data) >= 2500:
            break
    
   # if len(clean_data) >= 2500:
       # break
len(clean_data)
```
